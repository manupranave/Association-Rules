**Association Rule Mining:**

```python
!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
te
```

```python
te_ary=te.fit(transaction).transform(transaction)
te_ary
```

```python
data=DataFrame(te_ary,columns=te.columns_)
data.head()
```

```python
data.to_csv('data.csv',index=False)
```

```python
frequent_itemset=apriori(data,min_support=0.01,use_colnames=True)
frequent_itemset
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rule=association_rules(frequent_itemset,metric='confidence',min_threshold=0.5)
rule
```

**Analysis and Interpretation:**

```python
def format_itemset(itemset):
  return','.join(sorted(list(itemset)))
```

```python
rule['antecedents_formatted']=rule['antecedents'].apply(format_itemset)
rule['consequents_formatted']=rule['consequents'].apply(format_itemset)
rule
```

```python
print('Top 10 Association Rules :')
rule[['antecedents_formatted','consequents_formatted','support','confidence','lift']]
```

```python
antecedent=rule['antecedents_formatted'].value_counts().head(5)
consequent=rule['consequents_formatted'].value_counts().head(5)
```

```python
print('Top 5 Antecedents :')
antecedent
```

```python
print('Top 5 Consequents :')
consequent
```

**Interview Questions:**

1.	What is lift and why is it important in Association rules?

• Lift,a key concept in association rule mining,measures the strength of an association rule.

• It indicates how much greater the probability of the right-hand side(consequent)is in the presence of the left-hand side(antecedent)compared to its probability in the absence of the antecedent.

• A higher lift value represents a more significant association,making it more important in decision-making processes.

• Lift helps identifying the most relevent relationships between items,allowing bussinesses to optimize marketing stratagies,product recommendations,and customer segmentations.

2.	What is support and Confidence. How do you calculate them?

• Support and Confidence are fundamental concepts in decision-making and machine learning,particular in the classification and regression tasks.

• Support refers to the proportion of instances in the training dataset that belongs to certain class or category.

• Confidence represents the certainty or reliability of the model's prediction.

• Support calculated by counting the number of instances in each class and dividing by the total number of instances.

• Confidence calculated by dividing the number of transactions containing both itemsets by the number of transactions containing the first itemset.
